```python
from tkinter import *
import math
window=Tk()
til=window.title("CALCULATOR")

window.geometry("350x350")
window.resizable(0,0)
window.configure(cursor="dotbox")
#Set first frame of calculator for the entry part
frame1=Frame(window)
frame1.pack()

eqn=StringVar()
exprsn=""
# Define function for press(),clear() and eql()
def press(num):
    global exprsn
    exprsn=exprsn+str(num)
    eqn.set(exprsn)
def clear():
    global exprsn
    exprsn=""
    eqn.set(exprsn)
    
def eql():
    try:
        global exprsn
        total=str(eval(exprsn))
        eqn.set(total)
        exprsn=""
    except:
        eqn.set("Math Error")
        exprsn=""
#To Set Entry Widget for the entering the numbers
e1=Entry(frame1 ,bd =15,width=200,font="italics",bg='#FFFFD3',textvariable=eqn)
e1.pack(side="right",ipady=10)

#To Set 2nd frame for the button parts
frame2=Frame(window,bg='#CDCDAA')
frame2.pack()

#first row
clear = Button(frame2, text = "DEL", fg = "red", width = 10, height = 3, bd = 4,command=clear)
clear.grid(row=0,column=3)
divide = Button(frame2, text = "/", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("/"))
divide.grid(row=3,column=2)

# second row
seven = Button(frame2, text = "7", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("7")) 
seven.grid(row=0,column=0)
eight = Button(frame2, text = "8", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("8"))
eight.grid(row=0,column=1)
nine = Button(frame2, text = "9", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("9"))
nine.grid(row=0,column=2)
multiply = Button(frame2, text = "X", fg = "black", width = 10, height = 3, bd =4, command=lambda:press("*"))
multiply.grid(row=1,column=3)

# third row
four = Button(frame2, text = "4", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("4"))
four.grid(row=1,column=0)
five = Button(frame2, text = "5", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("5"))
five.grid(row=1,column=1)
six = Button(frame2, text = "6", fg = "black", width = 10, height = 3, bd = 4,  command=lambda:press("6"))
six.grid(row=1,column=2)
minus = Button(frame2, text = "-", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("-"))

# fourth row
one = Button(frame2, text = "1", fg = "black", width = 10, height = 3, bd =4 ,command=lambda:press("1"))
one.grid(row=2,column=0)
two = Button(frame2, text = "2", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("2"))
two.grid(row=2,column=1)
three = Button(frame2, text = "3", fg = "black", width = 10, height = 3, bd = 4, command=lambda:press("3"))
three.grid(row=2,column=2)
plus = Button(frame2, text = "+", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("+"))
plus.grid(row=2,column=3)

# fourth row
zero = Button(frame2, text = "0", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("0")) 
zero.grid(row=3,column=0)
point = Button(frame2, text = ".", fg = "black", width = 10, height = 3, bd = 4,command=lambda:press("."))
point.grid(row=3,column=1)
equals = Button(frame2, text = "=", fg = "black", width = 10, height = 3, bd = 4, command=eql)
equals.grid(row=3,column=3)



window.mainloop()
```
